## Age under 30

```python
data_30 = data[(data['age'] >= 30) & (data['age'] <= 40)]
age_n = '30 ~ 40歲'
os.mkdir(age_n)
```

### 年齡分布

```python
age = data_30['age']
title = '1 '+age_n + '  年齡分布圖'
plt.hist(age)
plt.title(title)
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 性別比

```python
gender = data_30['gender'].value_counts()
ssize = sum(gender)
plt.pie(gender, labels = gender.index, autopct = '%1.1f%%')
title = '2 '+age_n+'  性別比例'
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 身分

```python
identity = data_30['identity'].value_counts()
labels = list(zip(identity.index, identity))

plt.pie(identity, labels =labels, autopct= '%1.1f%%')
title = '3 '+age_n+'  身分比例'

ssize = sum(identity)
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 工作幾年

```python
work_y = data_30['years of work'].value_counts()
labels = list(zip(work_y.index, work_y))

plt.pie(work_y, labels = labels, autopct= '%1.1f%%')
title = '4 '+age_n + '  工作幾年'
ssize = sum(work_y)
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 月收入

```python
income = data_30['income (10k)'].values
income = income[~np.isnan(income)]
income_adj = []
for i in income:
    if i < 1000:
        income_adj.append(i)
    else:
        income_adj.append(i/10000)
        
income_adj.sort()
income_adj.pop()

plt.hist(income_adj)
plt.xlabel('收入(萬元)')
plt.ylabel('人次')

title = '5 '+age_n + '  收入分布圖'
plt.title(title)
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 對投資的興趣

```python
interest = data_30['interest'].value_counts()
plt.bar(interest.index, interest)
plt.xlabel('分數')
plt.ylabel('人數')

title = '6 '+age_n + '  投資興趣分布圖'
plt.title(title)
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 掌握程度

### 投資習慣

```python
habit = data_30['habit'].value_counts()
labels = list(zip(habit.index, habit))
plt.pie(habit, labels = labels, autopct = '%1.1f%%')

title = '8 '+ age_n + '  是否有投資習慣'
ssize = sum(habit)
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 遇到的障礙

```python
obs = list(data_30['obstacles'].dropna())
ssize = len(obs)
obs = ';'.join(obs)
obs = obs.split(';')

obs = pd.DataFrame(obs)
obs = obs[0].value_counts()
plt.barh(obs.index, obs/ssize, color = 'red')

title = '9 '+age_n+ '  投資障礙 (樣本數' + str(ssize) + ')'
plt.title(title)
plt.xlabel('比例')
address = './'+ age_n + '/'+ title
plt.savefig(address, bbox_inches='tight')
plt.show()
```

### 傾向學習管道

```python
tend_l = data_30['learn(tend)'].dropna()
ssize = len(tend_l)
tend_l_l = list(tend_l)
tend_l_s = ";".join(tend_l_l)

tend_l = tend_l_s.split(';')
tend_l_pd = pd.DataFrame(tend_l)
tend_l_count = tend_l_pd[0].value_counts()

plt.barh(tend_l_count.index, tend_l_count/ssize)
title = '91 '+ age_n + '傾向透過的學習管道 (樣本數' + str(ssize)+')'
plt.title(title)
address = './'+ age_n + '/'+ title
plt.savefig(address, bbox_inches='tight')
plt.show()
```

### 傾向學習網路管道

```python
tend_l = data_30['internet learn(tend)'].dropna()
ssize = len(tend_l)
tend_l_l = list(tend_l)
tend_l_s = ";".join(tend_l_l)

tend_l = tend_l_s.split(';')
tend_l_pd = pd.DataFrame(tend_l)
tend_l_count = tend_l_pd[0].value_counts()

plt.barh(tend_l_count.index, tend_l_count/ssize)
title = '91 '+age_n + '傾向透過的網路管道 (樣本數' + str(ssize)+')'
plt.title(title)

address = './'+ age_n + '/'+ title
plt.savefig(address, bbox_inches='tight')
plt.show()
```

### 投資標的 第一名

```python
targets = ['IT(stocks)', 'IT(mutual fund)', 'IT(foreign exchange)', 
           'IT(ETF)', 'IT(fixed deposit)', 'IT(saving insurance)', 'IT(future & options)']
targets_c = ['股票','基金','外匯','ETF','定存','儲蓄險','期貨 & 選擇權']

## for all
target_counts = []
for i in targets:
    tempt = data_30[i].value_counts().sort_index()
    if tempt.index[0] != '第 1' :
        target_counts.append(0)
        continue
    tempt = tempt[0]
    target_counts.append(tempt)

plt.pie(target_counts, labels = targets_c, autopct = '%0.2f%%')
title = age_n +'  使用投資標的排名 1 st'

ssize = sum(target_counts)
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()

## for male
target_counts = []
for i in targets:
    tempt = data_30[data_30['gender'] == '男'][i].value_counts().sort_index()
    if tempt.index[0] != '第 1' :
        target_counts.append(0)
        continue
    tempt = tempt[0]
    target_counts.append(tempt)

plt.pie(target_counts, labels = targets_c, autopct = '%0.2f%%')
title = age_n +'  使用投資標的排名 1 st- 男性'
ssize = sum(target_counts)
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()


## for female
target_counts = []
for i in targets:
    tempt = data_30[data_30['gender'] == '女'][i].value_counts().sort_index()
    if tempt.index[0] != '第 1' :
        target_counts.append(0)
        continue
    tempt = tempt[0]
    target_counts.append(tempt)

plt.pie(target_counts, autopct = '%0.2f%%', labels = targets_c)
title = age_n +'  使用投資標的排名 1 st- 女性'
ssize = sum(target_counts)
plt.title(title+'\n樣本數：'+str(ssize))
address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 每周關注的時間

```python
spend = data_30['spent time(hour)'].value_counts().sort_index()
data_len = len(data_30['spent time(hour)'].dropna())
length = [1.5, 3.5, 6, 8, 0]

total = 0
for i in range (len(spend)):
    total += length[i]*spend[i]

mean = round(total/data_len, 2)
# plot
plt.pie(spend, labels = spend.index, autopct = '%0.1f%%')
total = sum(spend)
plt.title('mu = ' + str(mean)+ ', 樣本數: '+ str(total), fontsize = 8)



title = age_n + '  每周關注的時間'
plt.suptitle(title)


address = './'+ age_n + '/'+ title
plt.savefig(address)
plt.show()
```

### 傾向學習
